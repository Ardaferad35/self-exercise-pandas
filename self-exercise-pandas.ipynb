```python
olymics_data['height_category'] = olymics_data['height_cm'].apply(
    lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall') if pd.notna(x) else 'NaN'
)
```

```python
olymics_data.head()
```

```python
def categorize_athlete(row):
    if pd.isna(row['height_cm']) or pd.isna(row['weight_kg']):
        return 'Unknown'
    elif row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 and row['weight_kg'] <= 80:
        return 'Middleweight'
    else:
        return 'Heavyweight'



olymics_data['Category'] = olymics_data.apply(categorize_athlete, axis=1)
```
